## Restructure Data

```python
weather = pd.read_csv('data\weather.csv')
weather.head()
```

```python
#days into single column we use melt=columns into rows
```

```python
a = pd.melt(weather,
        id_vars=['year','month','element'],
        var_name ='day',
       value_name='temp')
a.head(10)
```

```python
weather.shape
```

```python
a.shape
```

```python
##max and min as sep colimns we pivot=rows into columns
```

```python
b= a.pivot_table(index=['year','month','day'],
             columns='element',
             values='temp').reset_index()
b.head(10)
```

### Saving data into a file

```python
b.to_csv('output/weather_tidy.csv')
```

## Quering

```python
b.loc[(b.month==3)&(b.day=='day1')]
```

## custom methods

```python
def mean_func(x):
    return sum(x)/len(x)
```

```python
b[['min','max']].apply(mean_func,axis=1)
```

## Merging 2 dataframes

```python
users = pd.DataFrame({'id':[1,2,3],'name':["a","b","c"]})
users
```

```python
msgs = pd.DataFrame({'userid':[1,3,2],'msg':["hello","bye","hi"]})
msgs
```

```python
users.merge(msgs, left_on='id',right_on='userid')
```
