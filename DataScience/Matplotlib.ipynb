```python
import numpy as np
```

```python
x = np.arange(1,10)
y = x**2
plt.title('Squares')
plt.xlabel('x-values')
plt.ylabel('square-values')
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(-10,100)
plt.show()
```

```python
x = np.arange(-10,20)
y = x**2
x2= np.arange(-10,20)
y2 = x2+50
plt.plot(x,y,'y*',label='Squares')
plt.plot(x2,y2,label='linear')
plt.legend()
plt.grid()
plt.text(7.5,60,'hello')
plt.show()
```

```python
plt.plot(x,y,'b',x2,y2,'ro')
plt.show()
```

```python
x = np.arange(1,101)
y = np.log(x)
plt.plot(x,y)
plt.show()
```

```python
import math
x = np.arange(1,101)
# y = [math.log(i) for i in x]
# y = list(map(lambda i:math.log(i),x))
y= np.vectorize(math.log)(x)

plt.plot(x,y)
plt.show
```

### Bar graph

```python
plt.bar([1,2,3],[10,20,5],width=0.2,color='b',label='crime rate')
plt.bar([1.5,2.5,3.5],[5,10,15],width=.2,color='r',label='cost')
plt.legend()
plt.show()
```

### Histogram

```python
vals = [1,2,3,2,1,2,3,2,1,2,1,2,1]
plt.hist(vals)
plt.show()
```

### Normal Distribution

```python
mu = 100
sigma = 15
s = np.random.normal(mu,sigma,1000)
s
```

```python
count,bins,patches = plt.hist(s,20)
plt.show()
```

### Scatter Plot

```python
x = np.random.rand(20)
y = np.random.rand(20)

x2 = np.random.rand(20)
y2 = np.random.rand(20)

plt.scatter(x,y)
plt.scatter(x2,y2)
plt.show()
```

### Pie Chart

```python
plt.pie([10,20,30,40],
       labels=['Law','Education','Health','Defense'],
       colors=['c','m','r','b'],
       startangle=90,
       shadow=True,
       explode =(0,0.2,0,0))
plt.show()
```

```python
import math
import numpy
```

```python
sigmoid = lambda x:1/(1+math.exp(-x))
x = np.arange(-10,10)
y = [sigmoid(i) for i in x]
plt.plot(x,y)
plt.show()
```
